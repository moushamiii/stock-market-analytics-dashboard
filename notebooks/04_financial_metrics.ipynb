# Notebook 4: Company Fundamentals

## Objective

The purpose of this notebook is to collect and explore key financial fundamentals for five major technology companies using the Yahoo Finance API.

While the historical stock price dataset provides information on how each company's share price has changed over time, it does not explain why certain companies may have outperformed others. To gain deeper insight, this notebook gathers company-level financial metrics such as market capitalization, price-to-earnings ratio, earnings per share, revenue, profit margin, dividend yield, and return on equity.

These metrics will be combined with the historical stock price data in later stages of the project to support a more comprehensive comparison of company performance and to provide context for the investment analysis.

### Business Question

**What financial characteristics differentiate the five companies, and how might these metrics help explain differences in stock performance?**

```python
import yfinance as yf
import pandas as pd

tickers = ["AAPL","MSFT","NVDA","AMZN","GOOGL"]

financial_data = []

for ticker in tickers:

    company = yf.Ticker(ticker)

    info = company.info

    financial_data.append({

        "Ticker": ticker,

        "Company": info.get("longName"),

        "Sector": info.get("sector"),

        "Industry": info.get("industry"),

        "Market Cap": info.get("marketCap"),

        "PE Ratio": info.get("trailingPE"),

        "Forward PE": info.get("forwardPE"),

        "EPS": info.get("trailingEps"),

        "Dividend Yield": info.get("dividendYield"),

        "Revenue": info.get("totalRevenue"),

        "Net Income": info.get("netIncomeToCommon"),

        "Profit Margin": info.get("profitMargins"),

        "Return on Equity": info.get("returnOnEquity")

    })
```

```python
financials = pd.DataFrame(financial_data)

financials

financials.info()

financials.describe()

financials.to_csv(
    "../data/processed/company_fundamentals.csv",
    index=False
)

```
